Import Libraries and Read

```python
#Import CSV

import pandas as pd

BTC_df = pd.read_csv("C:/Users/dalmi/Documents/España/Data Analytics - IronHack/Quests/First_Project/First-Project/datos_BTC")
BTC_df
```

Rename Columns

```python
BTC_df2 = BTC_df.copy()
BTC_df2 = BTC_df2.rename(columns={"timestamp":"date"})
BTC_df2
```

Sort Columns ---> Ascending Date

```python
BTC_df2 = BTC_df2.sort_values(by="date").reset_index(drop=True)
BTC_df2
```

Format Columns ---> Date (d/m/y)

```python
BTC_df2["date"] = pd.to_datetime(BTC_df2["date"], unit="s").dt.strftime("%d/%m/%Y")
BTC_df2
```

Add Columns ---> Months, Years

```python
BTC_df2["month"] = BTC_df2["date"].apply(lambda x:x[3:5])
BTC_df2["year"] = BTC_df2["date"].apply(lambda x:x[6:10])
BTC_df2
```

Format Columns ---> Rename Months

```python
# Months dictionaries

meses_dictiionary = {
    "01": "Jan",
    "02": "Feb",
    "03": "Mar",
    "04": "Apr",
    "05": "May",
    "06": "Jun",
    "07": "Jul",
    "08": "Aug",
    "09": "Sep",
    "10": "Oct",
    "11": "Nov",
    "12": "Dec"
}
    
#Implement dictionary in columns months

BTC_df2["month"] = BTC_df2["month"].replace(meses_dictiionary)
BTC_df2
```

Format Columns ---> Change Order

```python
BTC_df2 = BTC_df2[["year","month","date","price"]]
BTC_df2
```

SAVE DATAFRAME to CSV

```python
BTC_df2.to_csv("datos_BTC_cleaned", header=True, index=False)
```
