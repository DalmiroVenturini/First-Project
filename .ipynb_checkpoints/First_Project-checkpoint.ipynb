```python
#Importo el CSV para trabajar el archivo
import pandas as pd

BTC_df = pd.read_csv("C:/Users/dalmi/Documents/España/Data Analytics - IronHack/Quests/First_Project/First-Project/datos_BTC")
BTC_df
```

```python
#Cambiar nombres de columnas

BTC_df2 = BTC_df.copy()
BTC_df2 = BTC_df2.rename(columns={"timestamp":"date"})
BTC_df2
```

```python
#Ordenar por fecha creciente

BTC_df2 = BTC_df2.sort_values(by="date").reset_index(drop=True)
BTC_df2
```

```python
#Convertir el formato de las fechas de "timestamps" a "%d/%m/%Y"

BTC_df2["date"] = pd.to_datetime(BTC_df2["date"], unit="s").dt.strftime("%d/%m/%Y")
BTC_df2
```

```python
#Agrego una columna para los meses y una para los años
#Creo la columna con meses y la agrego al DataFrame

BTC_df2["month"] = BTC_df2["date"].apply(lambda x:x[3:5])
BTC_df2["year"] = BTC_df2["date"].apply(lambda x:x[6:10])
BTC_df2
```

```python
#Reemplazar los meses por el nombre del mes
#Primero creo el diccionario de meses

meses_dictiionary = {
    "01": "Jan",
    "02": "Feb",
    "03": "Mar",
    "04": "Apr",
    "05": "May",
    "06": "Jun",
    "07": "Jul",
    "08": "Aug",
    "09": "Sep",
    "10": "Oct",
    "11": "Nov",
    "12": "Dec"
}
    
#Reemplazo la columna month por el diccionario

BTC_df2["month"] = BTC_df2["month"].replace(meses_dictiionary)
BTC_df2
```

```python
#Cambiar orden de columnas

BTC_df2 = BTC_df2[["year","month","date","price"]]
BTC_df2
```

```python
#Definir orden de miles con K en columna prices

BTC_df3 = BTC_df2.copy()

def format_thousands(value):
    return f"{value / 1000:.1f}K"

BTC_df3["price"] = BTC_df3["price"].apply(format_thousands)
BTC_df3
```

```python
#Creo un DataFrame que agrupe precio por mes tomando el último valor del mes

BTC_df3_month = BTC_df3.copy()

BTC_df3_month["date"] = pd.to_datetime(BTC_df3_month["date"])
BTC_df3_month.set_index("date", inplace=True)

#Filtrar el precio de cierre del último día de cada mes

BTC_df3_month = BTC_df3_month.resample("M").last().reset_index()
BTC_df3_month

#Eliminar columna de date

BTC_df3_month = BTC_df3_month.drop(columns=["date"])
BTC_df3_month
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

#Convertir la columna "date" a tipo datetime
#BTC_df3["date"] = pd.to_datetime(BTC_df3["date"], format='%d/%m/%Y')

# Graficar los precios
plt.figure(figsize=(12, 6))
plt.plot(BTC_df3["date"], BTC_df3["price"].str.replace("K", "").astype(float) * 1000, label="Bitcoin Price", color="blue")

#Formatear el eje x para mostrar solo meses y años
#plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%m %Y"))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=1))

# Rotar las etiquetas del eje x
plt.gcf().autofmt_xdate(rotation=90)

# Etiquetas y título
plt.title('Evolución del precio de Bitcoin')
plt.xlabel('Fecha')
plt.ylabel('Precio en USD')
plt.legend()
plt.grid()

# Mostrar el gráfico
plt.show()
```
