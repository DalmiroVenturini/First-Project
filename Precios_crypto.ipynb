```python
#Get the link

link_overview = "https://www.tradingview.com/markets/cryptocurrencies/"
link_coins = "https://www.tradingview.com/markets/cryptocurrencies/prices-all/"

response = requests.get(link_coins)
response.status_code # 200 status code means OK!
```
