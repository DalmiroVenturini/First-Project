Import Libraries and Read DataFrame

```python
import pandas as pd

# Import CSV

BTC_df2 = pd.read_csv("C:/Users/dalmi/Documents/España/Data Analytics - IronHack/Quests/First_Project/First-Project/datos_BTC_cleaned")
BTC_df2
```

BTC Closing Price by Month

```python
BTC_df2_month = BTC_df2.drop_duplicates(subset=["year","month"],keep="last")
BTC_df2_month
```

BTC Mean Price by Year

```python
price_mean = round(BTC_df2.groupby(by=["year"])["price"].agg(["mean"]))
price_mean
```

BTC Max Price by Year

```python
price_max =round(BTC_df2.groupby(by=["year"])["price"].agg(["max"]))
price_max
```

BTC Historical Price Plot (2019 - 2024)

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Graficar los precios
plt.figure(figsize=(12, 6)) #tamaño de gráfico: 12x6 pulgadas
plt.plot(BTC_df2["date"], BTC_df2["price"].astype(float) * 1000, label="Bitcoin Price", color="blue")

#Formatear el eje x para mostrar intervalos mensuales
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=1))

# Rotar las etiquetas del eje x
plt.gcf().autofmt_xdate(rotation=-85,ha="left")

# Etiquetas y título
plt.title('Evolución del precio de Bitcoin')
plt.xlabel('Fecha')
plt.ylabel('Precio en USD')
plt.legend()
plt.grid()

# Mostrar el gráfico
plt.show()
```
